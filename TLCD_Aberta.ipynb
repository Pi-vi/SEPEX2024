```python
import pandas as pd
```

```python
dados1 = pd.read_excel(r'ensaio_valvula_aberta.xlsx')
```

```python
dados1.describe
```

```python
tf = 3564
```

```python
tempo = pd.to_datetime(dados1.loc[13:tf,'Signal Name'])
```

```python
acc = dados1.loc[13:tf,'TimeHistory0026_Acc']
```

```python
import numpy as np
```

```python
tempo_s = np.zeros(tf-13)
for i in range(13,tf):
    tempo_s[i-13] = (tempo[i] - tempo[13]).total_seconds()
```

```python
acc_g = np.zeros(tf-13)
for i in range(13,tf):
    acc_g[i-13] = float(acc[i])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(dpi=300)
plt.plot(tempo_s,acc_g)
plt.xlabel('Tempo [s]')
plt.ylabel('AceleraÃ§Ã£o [g]')
plt.grid(True)
plt.savefig('Aberta_1.png')
plt.show()
```
